```python
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
stocks=['NOK','AMD','POET','PLUG','QS']
data=yf.download(stocks,start='2023-01-01',end='2025-01-29')
close_1=data['Close']

returns=close_1.pct_change().dropna()
fig,axs=plt.subplots(3,2,figsize=(16,8),gridspec_kw={'hspace':0.3,'wspace':0.2})

axs[0,0].plot(returns['NOK'])
axs[0,0].set_title('NOK')
axs[0,0].set_ylim([-0.5,0.5])

axs[0,1].plot(returns['QS'])
axs[0,1].set_title('QS')
axs[0,1].set_ylim([-0.5,0.5])

axs[1,0].plot(returns['PLUG'])
axs[1,0].set_title('PLUG')
axs[1,0].set_ylim([-0.5,0.5])

axs[1,1].plot(returns['POET'])
axs[1,1].set_title('POET')
axs[1,1].set_ylim([-0.5,0.5])

axs[2,0].plot(returns['AMD'])
axs[2,0].set_title('AMD')
axs[2,0].set_ylim([-0.5,0.5])

fig.delaxes(axs[2,1])
plt.show()
```

```python
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
stocks=['NOK','AMD','POET','PLUG','QS']
data=yf.download(stocks,start='2023-01-01',end='2025-01-29')
close_1=data['Close']

fig,axs=plt.subplots(3,2,figsize=(16,8),gridspec_kw={'hspace':0.2,'wspace':0.1})
returns=close_1.pct_change().dropna(axis=0)

axs[0,0].hist(returns['POET'],bins=50,range=(-0.2,0.2))
axs[0,0].set_title('POET')

axs[0,1].hist(returns['PLUG'],bins=50,range=(-0.2,0.2))
axs[0,1].set_title('PLUG')

axs[1,0].hist(returns['AMD'],bins=50,range=(-0.2,0.2))
axs[1,0].set_title('AMD')

axs[1,1].hist(returns['QS'],bins=50,range=(-0.2,0.2))
axs[1,1].set_title('QS')

axs[2,0].hist(returns['NOK'],bins=50,range=(-0.2,0.2))
axs[2,0].set_title('NOK')

fig.delaxes(axs[2,1])
plt.show()
```

```python
import pandas as pd
import yfinance as yf
from datetime import date, timedelta
from sklearn.linear_model import LinearRegression

today = date.today()
end_date = (today + timedelta(days=1)).strftime("%Y-%m-%d")
start_date = (today - timedelta(days=730)).strftime("%Y-%m-%d")

data = yf.download('AMD', start=start_date, end=end_date, progress=False)
data.columns = data.columns.droplevel(1)
data = data[['Close']]
data['Prediction'] = data['Close'].shift(-1)
data.dropna(inplace=True)

X = data[["Close"]]
y = data[['Prediction']]

model = LinearRegression()
model.fit(X, y)

last_price = data[['Close']].iloc[-1]
print("Last Price:", last_price.values[0])

next_day = model.predict(last_price.values.reshape(1, -1))
print("Prediction Next Day price:", next_day[0])
```
